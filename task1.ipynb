```python
# Task 1: Exploring and Visualizing the Iris Dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset with pandas (CSV source)
iris_csv_url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(iris_csv_url)

# Basic inspection
print("Shape:", df.shape)
print("\nColumn Names:", df.columns.tolist())
print("\nFirst 5 rows:\n", df.head())
print("\nDataset Info:")
df.info()
print("\nDescriptive Statistics:\n", df.describe())

# Visualization 1: Scatter plot (petal_length vs petal_width)
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species", palette="deep")
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.tight_layout()
plt.savefig("iris_scatter.png")
plt.show()

# Visualization 2: Histograms
df.drop(columns="species").hist(figsize=(10, 6), bins=20, color="steelblue", edgecolor="black")
plt.suptitle("Feature Distributions (Histograms)", fontsize=14)
plt.tight_layout()
plt.savefig("iris_histograms.png")
plt.show()

# Visualization 3: Box plots by species
plt.figure(figsize=(12, 5))
for i, col in enumerate(["sepal_length", "sepal_width", "petal_length", "petal_width"]):
    plt.subplot(1, 4, i + 1)
    ax = sns.boxplot(data=df, x="species", y=col, hue="species", palette="Set2", dodge=False)
    if ax.legend_ is not None:
        ax.legend_.remove()
    plt.title(col)
    plt.xticks(rotation=15)

plt.suptitle("Box Plots by Species", fontsize=13)
plt.tight_layout()
plt.savefig("iris_boxplots.png")
plt.show()

print("\nTask 1 complete. All plots saved.")
```
